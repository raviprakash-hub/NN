```python
import pandas as pd
data=pd.read_csv("minst-train.csv")
data
```

```python
features=data.columns[1:]

X=data[features]
y=data['label']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X/255,y,test_size=0.15,random_state=0)
```

```python
import numpy as np
from keras.utils import np_utils

print(np.unique(y_train,return_counts=True))
```

```python
n_classes=10
y_train=np_utils.to_categorical(y_train,n_classes)
```

```python
y_train
```

```python
y_test=np_utils.to_categorical(y_test,n_classes)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout

mnist_nn=Sequential()
```

```python
#hidden layer
mnist_nn.add(Dense(units=100,kernel_initializer='uniform',activation='relu',input_shape=(784,)))

#output layer
mnist_nn.add(Dense(units=10,kernel_initializer='uniform',activation='softmax'))

mnist_nn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

mnist_nn.fit(X_train,y_train,batch_size=64,epochs=20,validation_data=(X_test,y_test))
```

```python
#hidden layer
mnist_nn.add(Dense(units=100,kernel_initializer='uniform',activation='relu',input_shape=(784,)))

mnist_nn.add(Dropout(0.2))

#output layer
mnist_nn.add(Dense(units=10,kernel_initializer='uniform',activation='softmax'))

mnist_nn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

mnist_nn.fit(X_train,y_train,batch_size=64,epochs=20,validation_data=(X_test,y_test))
```

```python
y_pred=mnist_nn.predict(X_test)
y_pred
```

```python
y_pred=(y_pred >0.9)
y_pred
```
